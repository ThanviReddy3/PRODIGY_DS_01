```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(
    r"C:\Users\india\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_247957\API_SP.POP.TOTL_DS2_en_csv_v2_247957.csv",
    skiprows=4  )
```

```python
df
```

```python
df.drop(columns = ['Unnamed: 68','Indicator Name', 'Indicator Code'], inplace=True )
```

```python
df.isnull().sum()
```

```python
india = df[df['Country Name'] == 'India']  #lets take india data 
```

```python
years = india.columns[2:]
```

```python
population = india.iloc[:,2:].values.flatten().astype('int')
```

## Bar

```python
plt.figure(figsize=(14,12))
plt.bar(years, population)
plt.xticks(rotation=90)
plt.title('Population of india over years')
plt.xlabel('years')
plt.ylabel('population')
plt.show()
```

## Histogram

```python
population_series = pd.Series(population)
plt.figure(figsize=(14, 6))
population_series.plot(kind='hist', bins=20, edgecolor='black')
plt.title('Histogram of Population Data')
```
